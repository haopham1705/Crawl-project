```python
# Import library and packages for object
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from time import sleep
import pandas as pd
print("import successful")
```

```python
# Login into webpage
url_login = "https://careerbuilder.vn/"
driver    = webdriver.Chrome()
driver.get(url_login)
sleep(2)
```

```python
search_input =driver.find_element_by_xpath("/html/body/main/div[1]/section[2]/div/div/div[1]/form/div[1]/div[1]/div/input")
search_input.clear()
input_text = str(input())
search_input.send_keys(input_text)
search_input.send_keys(Keys.RETURN)
# btn_search = driver.find_element_by_xpath("/html/body/main/div[1]/section[2]/div/div/div[1]/form/div[4]/button")
sleep(2)
# print(btn_search) 
```

```python
next_btn = driver.find_element_by_xpath('/html/body/main/section[2]/div/div/div[1]/div[2]/div/div/div/div[2]/ul/li[6]/a')
next_btn.click()
```

```python
page_source = BeautifulSoup(driver.page_source, 'lxml')

job_object = []
next_btn = driver.find_element_by_xpath('/html/body/main/section[2]/div/div/div[1]/div[2]/div/div/div/div[2]/ul/li[6]/a')
next_btn.click()
sleep(2)
input_page = int(input("How many page you want to scape?"))
# next button for first page
# driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
# next_btn_first_page = driver.find_element_by_xpath('/html/body/main/section[2]/div/div/div[1]/div[2]/div/div/div/div[2]/ul/li[6]/a')
# next_btn_first_page.click()
list_jobs = page_source.find_all('div', class_='job-item')
for page in range(input_page):
    

    for job_item in list_jobs:
        dict_info = {
        "title": "",
        "company_name" :"",
        "salary" : "",
        "date_post" : "",
        }
        list_titles = job_item.find_all('div', class_='title')
        text_job = list_titles[0].text.strip()
#         print(text_job)
#         Company name
        list_companys = job_item.find_all('a', class_='company-name')
        company_name_item = list_companys[0].text.strip()
#         print(company_name_item)
#         Salary
        list_salary = job_item.find_all('div', class_='salary')
        salary_item = list_salary[0].text.strip()
#         print(salary_item)
#         Time
        list_date = job_item.find_all('time')
        date_item = list_date[0].text.strip()
#         print(date_item)
#         add info
        dict_info["title"] = text_job
        dict_info["company_name"] = company_name_item
        dict_info["salary"] = salary_item
        dict_info["date_post"] = date_item
        
        print(dict_info, "dict-info after scaping")
        job_object.append(dict_info)
    
#     driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    sleep(2)
    next_btn = driver.find_element_by_xpath('/html/body/main/section[2]/div/div/div[1]/div[2]/div/div/div/div[2]/ul/li[7]/a')
    next_btn.click()
    sleep(2)
print('finish scraping')
```

```python
def execute_page():
    input_page = int(input("How many page you want to scape?"))
    # next button for first page
    # driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    # next_btn_first_page = driver.find_element_by_xpath('/html/body/main/section[2]/div/div/div[1]/div[2]/div/div/div/div[2]/ul/li[6]/a')
    # next_btn_first_page.click()

    for page in range(input_page):

        list_jobs = page_source.find_all('div', class_='job-item') 
        for job_item in list_jobs:
            dict_info = {
            "title": "",
            "company_name" :"",
            "salary" : "",
            "date_post" : "",
            }
            list_titles = job_item.find_all('div', class_='title')
            text_job = list_titles[0].text.strip()
    #         print(text_job)
    #         Company name
            list_companys = job_item.find_all('a', class_='company-name')
            company_name_item = list_companys[0].text.strip()
    #         print(company_name_item)
    #         Salary
            list_salary = job_item.find_all('div', class_='salary')
            salary_item = list_salary[0].text.strip()
    #         print(salary_item)
    #         Time
            list_date = job_item.find_all('time')
            date_item = list_date[0].text.strip()
    #         print(date_item)
    #         add info
            dict_info["title"] = text_job
            dict_info["company_name"] = company_name_item
            dict_info["salary"] = salary_item
            dict_info["date_post"] = date_item

            #print(dict_info, "dict-info after scaping")
            job_object.append(dict_info)

    #     driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        sleep(2)
        next_btn = driver.find_element_by_xpath('/html/body/main/section[2]/div/div/div[1]/div[2]/div/div/div/div[2]/ul/li[7]/a')
        next_btn.click()
        sleep(2)
```
