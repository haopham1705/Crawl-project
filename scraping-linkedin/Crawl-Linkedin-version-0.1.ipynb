```python
#import package
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
from bs4 import BeautifulSoup
import csv

print('finish importing package')
```

```python
#open URL and login
driver = webdriver.Chrome()
url = 'https://www.linkedin.com/login'
driver.get(url)
sleep(2)

credential = open('login_credential.txt')
line = credential.readlines()
username = line[0]
password = line[1]
print('Finished im porting the login credentials')
sleep(2)

email_field = driver.find_element_by_id('username')
email_field.send_keys(username)
sleep(3)

password_field = driver.find_element_by_name('session_password')
password_field.send_keys(password)
sleep(2)

signin_field = driver.find_element_by_xpath('//*[@id="organic-div"]/form/div[3]/button')
signin_field.click()
sleep(3)
print('Step 1: Login to Linkedin')
```

```python
# Search for the profile we want to crawl 
search_field = driver.find_element_by_xpath('//*[@id="ember17"]/input') 
search_query = input('What skill do you want to scrape? ')
search_field.send_keys(search_query) 
search_field.send_keys(Keys.RETURN) 
print('Step 2: Search for profiles')
```

```python
# filter profile
click_see_all = driver.find_element_by_xpath('//*[@id="main"]/div/div/div[1]/div/a')
click_see_all.click()
sleep(2)
```

```python
# scape the profile URLs
def GetURL():
    soup = BeautifulSoup(driver.page_source)
    profiles = soup.find_all('a', class_ = 'app-aware-link')
    all_profile_URL = []
    for profile in profiles:
        profile_URL = profile.get('href') 
        if profile_URL not in all_profile_URL:
            all_profile_URL.append(profile_URL)
    return all_profile_URL

input_page = int(input('How many pages you want to scrape: '))
URLs_all_page = []
for page in range(input_page):
    URLs_one_page = GetURL()
    sleep(2)
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    sleep(3)
    next_button = driver.find_element_by_class_name("artdeco-pagination__button--next")
    driver.execute_script("arguments[0].click();", next_button)
    URLs_all_page = URLs_all_page + URLs_one_page
    sleep(2)
print('Step 3: Scrape the URLs')
```

```python
# Scrape the data of 1 Linkedin profile, and write the data to a .CSV file
with open('output.csv', 'w',  newline = '') as file_output:
    headers = ['Name', 'Location', 'Job Title', 'URL']
    writer = csv.DictWriter(file_output, delimiter=',', lineterminator='\n',fieldnames=headers)
    writer.writeheader()
    for linkedin_URL in URLs_all_page:
        sleep(3)
        driver.get(linkedin_URL)
        print('- Accessing profile: ', linkedin_URL)
        check_link_profile = "https://www.linkedin.com/in/"
        soup = BeautifulSoup(driver.page_source, "html.parser")
        if check_link_profile in linkedin_URL:
            sleep(1)
            try:
                info_div = soup.find('div',{'class': 'pv-text-details__left-panel mr5'})
                info_profile = info_div.find_all('div')
                name = info_profile[0].find('h1').get_text().strip()
                print(name)
                title = info_profile[1].get_text().strip()
                print(title)
                location = info_profile[2].find('span').get_text().strip()
                print(location)
                type(location)
                writer.writerow({headers[0]:name, headers[1]:location, headers[2]:title, headers[3]:linkedin_URL})
                print('\n')
            except NameError:
                print('Variable is not defined')
            except:
                print('Some else went wrong')
            else:
                print('Profile OK')
print('Completed crape!')
```
