```python
# Create a function

def execute_page(page,key):

    # Soup the current page
    soup = BeautifulSoup(driver.page_source,'lxml')
    
    # Create a empty dataframe
    df_pie = pd.DataFrame(None,columns = ['job_title'   ,
                                         'job_link'     ,
                                         'company_name' ,
                                         'job_exp'      ,
                                         'location'     ,
                                         'job_descript' ,
                                         'job_skill'    ,
                                         'date_post'    ,
                                         'page'         ,
                                         'keyword'
                                        ])
    # Find all info
    infos = soup.find_all('li',class_ = 'clearfix job-bx wht-shd-bx')
    print("found: {} jobs".format(len(infos)))
    
    for info in infos:
        #try to get info if can not pass
        try:
            # Extract info in a div
            job_title    = info.find('a').text.strip()
            job_link     = info.find('a').get('href')
            company_name = info.find('h3').text.strip()
            job_exp      = info.find('ul',class_ = 'top-jd-dtl clearfix').find_all('li')[0].text
            location     = info.find('ul',class_= 'top-jd-dtl clearfix').find_all('li')[-1].text
            job_descript = info.find_all('li')[2].text
            job_skill    = info.find_all('li')[-1].text.strip()
            date_post    = info.find_all('span')[-1].text

            # Create dataframe
            df = pd.DataFrame({'job_title'      : [job_title],
                                 'job_link'     : [job_link] ,
                                 'company_name' : [company_name],
                                 'job_exp'      : [job_exp] ,
                                 'location'     : [location],
                                 'job_descript' : [job_descript],
                                 'job_skill'    : [job_skill],
                                 'date_post'    : [date_post],
                                 'page'         : [page],
                                 'keyword'      : [key]
                                 })

            # concatenate into dataframe
            df_pie = pd.concat([df_pie,df])
        
        except:
            print('Can not get info this one!,Move to the next')
            pass
        
    return df_pie
```

```python
# Step1: import libraries and packages for object
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys 
import time
from bs4 import BeautifulSoup
import pandas as pd
import datetime

# Say something welcome
print("\n Hello, i am crawler TimeJobs v1 \n")
print("\n Nice to meet you \n")

# Field the keyword into the 
keyword      = input("What is your keyword? ")
page_no      = int(input("How many page you want to drawl? "))
data_id      = input("What is your data id? ")

# Login into webpage
url_page = 'https://www.timesjobs.com/candidate/job-search.html?from=submit&actualTxtKeywords=java&searchBy=0&rdoOperator=OR&searchType=personalizedSearch&luceneResultSize=25&postWeek=60&txtKeywords=java&pDate=I&sequence=1&startPage=1'
driver = webdriver.Chrome()

l       = url_page.split('&')
l[1]    = 'actualTxtKeywords=' + keyword

url = ''
for i in l:
    url = url+i+'&'
url = url[:-1]

driver.get(url)
time.sleep(15)

try:
    button_hide = driver.find_element_by_id('closeSpanId')
    button_hide.click()
except: pass

for i in range(1,page_no + 1):
    
    print("i: ",i)
    url   = driver.current_url   
    # Split the url via
    l     = url.split('&')
    # Change the page
    l[-2] = 'sequence=' + str(i)
    # Turn list into a string
    t     = ''
    for j in l:
        t = t + j + '&'
    # Replace the current page
    t = t[:-1]
    print(t)
    
    driver.get(t)
    
    df_i = execute_page(i,keyword)
    
    df_global = pd.concat([df_global,df_i])
    time.sleep(2)
    
    df_global.reset_index()
    
    date = datetime.datetime.now()
    yy = date.strftime("%Y")[-2:]
    mm = date.strftime("%m")
    dd = date.strftime("%d")
    ddmmyy =dd + mm + yy
    
    df_global.to_excel('timejobs_'+ddmmyy+'_'+data_id+'.xlsx')
    print("\n Exported! page {} \n".format(i))
    time.sleep(2)
    
print("\n Finished! Thank for using me :) \n")
```
