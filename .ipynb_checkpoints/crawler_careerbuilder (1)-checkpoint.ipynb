```python
# Step11: Complete your code

# Step1: import libraries and packages for object
from selenium import webdriver 
from selenium.webdriver.common.keys import Keys 
import time
from bs4 import BeautifulSoup
import pandas as pd
import datetime

# Step9: create a function
def execute_page(page,keyword):
    #Soup current_page
    soup = BeautifulSoup(driver.page_source,'lxml')
    
    #Create a empty dataframe
    df_big = pd.DataFrame(None,columns = ['company_name',
                                        'company_img',
                                        'job_link',
                                        'job_title',
                                        'salary',
                                        'location',
                                        'date',
                                        'page',
                                        'keyword'])
    
    # Step8: execute a page find all from
    div_infos = soup.find_all('div',class_ = "job-item")
    print("found: {} jobs".format(len(div_infos)))

    for div_info in div_infos:
        # Exctact info in a div
        company_name = div_info.find('a',class_ = 'company-name').text   
        company_img  = div_info.find('img').get('src') 
        job_link     = div_info.find('a',class_ = 'job_link').get('href')  
        job_title    = div_info.find('a',class_ = 'job_link').get('title')  
        salary       = div_info.find('div',class_ = "salary").text     
        location     = div_info.find('li').text
        date         = div_info.find('time').text

        # Create dataframe
        df = pd.DataFrame({'company_name': [company_name],
                            'company_img': [company_img],
                            'job_link'   : [job_link], 
                            'job_title'  : [job_title],
                            'salary'     : [salary],
                            'location'   : [location],
                            'date'       : [date],
                            'page'       : [page],
                            'keyword'    : [keyword]
                      })

        # Concate into dataframe
        df_big = pd.concat([df_big,df])
    
    return df_big

# Say something welcome
print("\n Hello, i am CareerBuilder v1 \n")
print("\n Nice to meet you \n")

# Field the keyword into the 
keyword      = input("What is your keyword? ")
page_no      = int(input("How many page you want to drawl? ")) 

# Login into webpage
url_page  = "https://careerbuilder.vn/"
driver    = webdriver.Chrome() # Open the driver of Chrome
driver.get(url_page) # go in to the url_link
time.sleep(2) # Hold abit for webpage render

# Step5: add the input and find result
# Find the element
search_field = driver.find_element_by_id('keyword') 
# enter keyword
search_field.send_keys(keyword)
# Hold a litlebit
time.sleep(1)
# Press Enter
search_field.send_keys(Keys.RETURN)

# Step7: create a dataframe
df_global = pd.DataFrame(None,columns = ['company_name',
                                        'company_img'  ,
                                        'job_link'     ,
                                        'job_title'    ,
                                        'salary'       ,
                                        'location'     ,
                                        'date'         ,
                                        'page'         ,
                                        'keyword'])

for i in range(1,page_no+1):
    
    # Execute the current page
    df_i = execute_page(i,keyword)
    
    # Concatenate data
    df_global = pd.concat([df_global,df_i])
    
    # Goto next page
    button_next = driver.find_element_by_xpath('/html/body/main/section[2]/div/div/div[1]/div[2]/div/div/div/div[2]/ul/li[6]/a')
    button_next.click()
    
    time.sleep(3)
    
    df_global.reset_index()
    
    date = datetime.datetime.now()
    yy = date.strftime("%Y")[-2:]
    mm = date.strftime("%m")
    dd = date.strftime("%d")
    ddmmyy =dd + mm + yy
    
    df_global.to_excel('careerbuilder_'+ddmmyy+'_1.xlsx')
    
print("\n Finished! Thank for using me :) \n")
```
